```python
def h_distance(X,Y):
    hd = np.max([directed_hausdorff(X,Y)[0], directed_hausdorff(Y,X)[0]])
    return hd/((np.sum(np.linalg.norm(X, axis=1)) + np.sum(np.linalg.norm(X, axis=1)))/(X.shape[0] + Y.shape[0]))
```
